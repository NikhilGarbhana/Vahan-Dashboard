```python
def table_data():
    rows=[]
    vahan_soup = BeautifulSoup(driver.page_source, 'html5lib')

    table_body = vahan_soup.find('tbody', id='groupingTable_data')
    for i in table_body.find_all('tr'):
        row = []
        for j in i.find_all('td'):
            title = j.text
            row.append(title)
        rows.append(row)

    return rows
```
